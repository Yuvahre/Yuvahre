```python
# Split the 'DATE' column into 'DATE' and 'TIME' columns
hourly['TIME'] = hourly['DATE'].str.split('T', expand=True)[1]
hourly['DATE'] = hourly['DATE'].str.split('T', expand=True)[0]
daily['TIME'] = daily['DATE'].str.split('T', expand=True)[1]
daily['DATE'] = daily['DATE'].str.split('T', expand=True)[0]
monthly['TIME'] = monthly['DATE'].str.split('T', expand=True)[1]
monthly['DATE'] = monthly['DATE'].str.split('T', expand=True)[0]
three_hour['TIME'] = three_hour['DATE'].str.split('T', expand=True)[1]
three_hour['DATE'] = three_hour['DATE'].str.split('T', expand=True)[0]
```

```python
hourly
```

```python
merged_data = pd.merge(hourly, daily, on=['STATION', 'DATE'], how='outer')
merged_data = pd.merge(merged_data, monthly, on=['STATION', 'DATE'], how='outer')
merged_data = pd.merge(merged_data, three_hour, on=['STATION', 'DATE'], how='outer')
```
